````python
url = 'https://www.top500.org/list/2014/11/?page=1'

link_list = []
name_list = []
res = requests.get(url).content
soup = BeautifulSoup(res,'html.parser')
#rank = element.find_all('td')[0].text
#name = element.find_all('td')[2].text.split()[0]
#description = ' '.join(element.find_all('a')[1].text.split()[2:])
table_tr = soup.select('table > tr')
for element in table_tr:
    name = [e.text for e in element.find_all('b') if element.find_all('b')]
    name_list.append(''.join(name))
name_list
    
    
````
